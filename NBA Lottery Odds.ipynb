```python
import pandas as pd
import numpy as np

chances = [114., 113., 112., 111.,  99.,  89.,  79.,  69.,  59.,  49.,  39., 29.,  19.,   9.,   6.,   4.]

num_teams = 16
seeds = np.arange(1, num_teams+1)
num_chances = np.array(chances)
rows = np.stack((seeds, num_chances), axis=-1)
df = pd.DataFrame(rows, columns = ['Seed', 'Chances'])


first = []
secnd = []
thrd = []
frth = []
ffth = []

old_seed2 = 1
old_seed3 = 1
old_seed4 = 1
old_seed5 = 1

temp_sec = 0
temp_third = 0
temp_fourth = 0
temp_fifth = 0

# GET ODDS ON FIRST PICK
for chance in chances:
    first.append(chance/1000)

for seed in seeds:
    print(seed)
    for seed2 in seeds:
        picks = [seed]
        
        # GET ODDS ON SECOND PICK
        if seed2 not in picks:
            p = (df.iloc[seed-1]['Chances']/1000.)*(df.iloc[seed2-1]['Chances']/ (1000. - df.iloc[seed2-1]['Chances']))
            if seed > old_seed2:
                secnd.append(temp_sec)
                temp_sec = 0
            temp_sec = temp_sec + p
            old_seed2 = seed
            
            # GET ODDS ON THIRD PICK
            for seed3 in seeds:
                picks = [seed, seed2]
                if seed3 not in picks:
                    p2 = (df.iloc[seed3-1]['Chances']/ (1000. - df.iloc[seed3-1]['Chances'] - df.iloc[seed2-1]['Chances']))
                    if seed > old_seed3:
                        thrd.append(temp_third)
                        temp_third = 0
                    temp_third = temp_third + (p2*p)
                    old_seed3 = seed
                    
                    # GET ODDS ON FOURTH PICK
                    for seed4 in seeds:
                        picks = [seed, seed2, seed3]
                        if seed4 not in picks:
                            p3 = (df.iloc[seed4-1]['Chances']/ (1000. - df.iloc[seed4-1]['Chances'] - df.iloc[seed3-1]['Chances'] - df.iloc[seed2-1]['Chances']))
                            if seed > old_seed4:
                                frth.append(temp_fourth)
                                temp_fourth = 0
                            temp_fourth = temp_fourth + (p3*p2*p)
                            old_seed4 = seed
                            
                            # GET ODDS ON FIFTH PICK
                            for seed5 in seeds:
                                picks = [seed, seed2, seed3, seed4]
                                if seed5 not in picks:
                                    p4 = (df.iloc[seed5-1]['Chances']/ (1000. - df.iloc[seed5-1]['Chances'] - df.iloc[seed4-1]['Chances'] - df.iloc[seed3-1]['Chances'] - df.iloc[seed2-1]['Chances']))
                                    if seed > old_seed5:
                                        ffth.append(temp_fifth)
                                        temp_fifth = 0
                                    temp_fifth = temp_fifth + (p4*p3*p2*p)
                                    old_seed5 = seed
                                    picks = [seed, seed2, seed3, seed4, seed5]   

                                    
secnd.append(temp_sec)
thrd.append(temp_third)
frth.append(temp_fourth)
ffth.append(temp_fifth)
```
